```python
import pandas as pd
import numpy as np
```

```python
def get_data(size):
    df = pd.DataFrame()
    df['age'] = np.random.randint(0, 100, size)
    df['time_spent_in_bed'] = np.random.randint(0, 9, size)
    df['pct_sleeping'] = np.random.rand(size)
    df['fav_food'] = np.random.choice(['pizzza', 'tacos', 'ice-cream'], size)
    df['hate_food'] = np.random.choice(['broccoli', 'candy corn', 'ice-eggs'], size)
    dates = pd.date_range('2020-01-01', '2022-12-31')
    df['effectve_dt'] = np.random.choice(dates, size)
    df['term_dt'] = np.random.choice(dates, size)
    df['expiration_dt'] = np.random.choice(dates, size)
    return df
```

```python
df1 = get_data(100)
df1
```

```python
def set_dtypes(df):
    df['age'] = df['age'].astype('int16')
    df['time_spent_in_bed'] = df['time_spent_in_bed'].astype('int16')
    df['pct_sleeping'] = df['pct_sleeping'].astype('float16')
    df['fav_food'] = df['fav_food'].astype('category')
    df['hate_food'] = df['hate_food'].astype('category')
    return df
```

```python
df2 = set_dtypes(df1)
df2
```
